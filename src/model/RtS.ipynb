```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Data
Material = [1555, 29636, 59865, 90696, 152671]
Labour = [3640, 1820, 1514, 1181, 1228]
Energy = [10, 191, 385, 583, 980]
Shipping = [0, 3812, 7700, 11665, 19636]
Marketing = [2000, 10000, 10000, 20000, 20000]
Other = [0, 95, 231, 408, 780]
year = [2025, 2026, 2027, 2028, 2029]

# Create dataframe
df = pd.DataFrame({'Year': year, 'Other': Other, 'Marketing': Marketing, 'Shipping': Shipping, 'Energy': Energy, 'Labour': Labour, 'Material': Material})

# Display dataframe
print(df.head())

df.plot( 
    x = 'Year', 
    xlabel='Cost (EUR)',
    kind = 'barh', 
    stacked = True, 
    title = 'Cost over Year', 
    mark_right = True,
    colormap = 'tab20')
```
